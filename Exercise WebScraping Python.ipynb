Extrae todas las cadenas que comienzan con `'https://`. Tienes que extraer desde `'https://` hasta el final. Ej.: 'https://img.milanuncios.com/fg/3749/86/374986816_1.jpg?VersionId=eUgig6FRM4L8.a8VbwgAsPfHUxDoCudE'

```python
#Importamos las librerias que vamos a utilizar durante 

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# T

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'
}

#Llamamos atraves de la funcion .get el link de la pagina y guardarla en la variable page
page = requests.get('https://www.milanuncios.com/motos-de-carretera/duke-390.htm', headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')
# Guardamos la URL en la variable link
link =('https://www.milanuncios.com/motos-de-carretera/duke-390.htm')
```

```python
#Creamos el dataframe df1 y lo utilizé a interior del for donde empezé a evaluar al interior de empazé a recorrer 
#todos las 'a' que se encuentran en el link y utilizé la función split la cual nos permite ir troceando todos los 'href'
#que vaya encontrando y con un segundo for evalué si las 8 primeras caracteres del texto, si encuentra que son un link
#procederá a guardarlo en el df1 que hemos creado en la pocisión de la iteración en la que nos encontré, y utilizé un
#try para saltar la iteración en caso de que no encuentre ningun lonk.

df1=pd.DataFrame()
i=0
for link in soup.find_all('a'):
    try:
        x=link.get('href').split("'")
        for pos in x:
            if pos[0:8] == "https://":
                df1.loc[i,0]=pos
                i+=1
    except:
        continue
df1
```
